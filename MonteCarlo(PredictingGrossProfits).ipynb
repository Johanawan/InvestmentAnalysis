### Predict firms future gross profit

Requirements :
1. Expected Revenue
2. Expected COGS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170 #rev_m = revenue_mean #Expected revenue $170 million
rev_stdev = 20 #Standard deviation is $20 million 
iterations = 1000 #How many simulated observations
```

```python
#Produce 1000 simulations of future revenues
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
#Plot the 1000 simulations on a graph
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()

#Obvious all data points fall inbetween 150 to 190
```

```python
#COGS are money spent so starts with -
#We need to assign 60% of COGS to each data point in graph above
COGS = - (rev * np.random.normal(0.6,0.1)) #Assume COGS are 60% of Rev #Assume standard deviation is 10%

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show
```

```python
COGS.mean()
```

```python
COGS.std()
```

##### Computing Gross Profit

```python
Gross_Profit = rev + COGS #Add COGS because we made it a negative number in line 17
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show()
```
