```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb

tickers = ['MCD', '^GSPC']
data = pd.DataFrame()
for t in tickers:
    data[t] = wb.DataReader(t, data_source = 'yahoo', start= '2014-11-15', end= '2019-11-15')['Adj Close']
```

```python
sec_returns = np.log(data / data.shift(1))
```

```python
cov = sec_returns.cov() * 250 #calculate covariance annually
cov
```

We are interest in: 
1. cov of PG and S&P
2. variance of S&P

```python
cov_with_markets = cov.iloc[0,1] #iloc: selecting data by row numbers. loc: selecting data by labels
cov_with_markets
```

```python
market_var = sec_returns['^GSPC'].var() * 250
market_var
```

### Beta

                        BETApg = Cov(pg,s&p)/Variance(s&p)

```python
MCD_beta = cov_with_markets / market_var
MCD_beta
```

MCD is a defensive stock as B>0 but B<1 so it if the market were to fall, PG would fall less then S&P

#### Calculate the expected return of MCD (CAPM)

                            Rpg = Rf + BETAmcd(Rm - Rf)

```python
#MCD_er == MCD Expected Return
#Look up risk free return on Bloomberg (10yr Treasury Note)
MCD_er = 0.01816 + MCD_beta *0.05184
MCD_er
```

#### Sharp Ratio

                Sharpe = (Rpg - Rf) / SDpg

```python
sec_returns['MCD'].std() * 250 ** 0.5
```

```python
Sharpe = (MCD_er - 0.025) / (sec_returns['MCD'].std() * 250 ** 0.5)
Sharpe
```
